# Clases y atributos del ndarray

**ndarray** es una clase de arreglo de Numpy, también se regiere como **numpy.ndarray**. Los atributos son los siguientes:

## ndim

Se refiere al número de ejes(dimensiones) del arreglo o matriz. Es también llamado el rango de la matriz.

```python
np_city = np.array(["NYC","LA","Miami","Houston"])
```

```python
print(np_city)
```

```python
np_city.ndim
```

```python
np_city_with_state = np.array([["NYC","LA","Miami","Houston"],["NY","CA","FL","TX"]])
```

```python
print(np_city_with_state)
```

```python
np_city_with_state.ndim
```

## shape

Tiene como output una tupla de enteros que muestran el tamaño del arreglo en cada dimensión. La longitud del **shape tuple** es el rango o ndim

```python
#1 fila y 4 columnas
np_city.shape
```

```python
#2 filas y 4 columnas
np_city_with_state.shape
```

## size

Muestra el número total de elementos del array o matriz. Es igual al producto de elementos del **shape tuple**

```python
np_city.size
```

```python
np_city_with_state.size
```

## dtype

Es un objeto que describe el tupe de elementos en un arreglo o matriz

El parámetro dtype indica el tipo de datos que se almacenan en el array. En este caso, "<U7" significa que los elementos del array son cadenas de texto con una longitud máxima de 7 caracteres.

```python
np_city
```

```python
np_city_with_state
```

```python
np_city_with_state.dtype
```

# Copy and Views

## Simple Assignments

```python
NYC_Borough = np.array(["Manhattan","Bronx","Brooklyn","Staten Island","Queens"])
NYC_Borough
```

```python
Boroughs_in_NYC = NYC_Borough
Boroughs_in_NYC
```

```python
Boroughs_in_NYC is NYC_Borough
```

## View/Shallow copy

```python
Boroughs_in_NYC
```

```python
View_of_Boroughs_in_NYC = Boroughs_in_NYC.view()
```

```python
len(View_of_Boroughs_in_NYC)
```

```python
#Change value in view object
View_of_Boroughs_in_NYC[4] = "Central Park"
View_of_Boroughs_in_NYC
```

```python
#Original dataset changed
Boroughs_in_NYC
```

## Deep copy

```python
Copy_of_NYC_Borought = NYC_Borough.copy()
```

```python
Copy_of_NYC_Borought is NYC_Borough
```

```python
#Cambiando valor en la copia
Copy_of_NYC_Borought[4] = "Central Park"
```

```python
#Original dataset
NYC_Borough
```

```python
Copy_of_NYC_Borought
```

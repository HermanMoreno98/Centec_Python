## Subplots

To create multiple plots within the same figure

.add_subplot(x,y,z)

- x = How many rows you want to add to the figure
- y = Hoy many columns you want to add
- z = The number of the plot that you want to add

```python
# Figure with 1 row and 2 columns
fig = plt.figure()

names = ["A","B","C"]
values = [19,50,29]
values_2 = [48,19,41]

ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.bar(names, values)
ax2.bar(names, values_2)
plt.show()
```

```python
# Figure with 2 rows and 1 column
fig = plt.figure()

names = ["A","B","C"]
values = [19,50,29]
values_2 = [48,19,41]

ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax.bar(names, values)
ax2.bar(names, values_2)
plt.show()
```

## Changing Figure Size

**figsize()**

The parametres is measured in "inches" (pulgadas)

figsize(8,6) -> width:8. height:6

```python
names = ["A","B","C"]
values = [19,50,29]
values_2 = [48,19,41]

fig = plt.figure(figsize=(20.0,15.0))

# Adds subplot on position 1
ax = fig.add_subplot(1,2,1)
# Adds subplot on position 2
ax2 = fig.add_subplot(1,2,2)

ax.bar(names,values)
ax2.bar(names,values_2)
plt.show()
```

```python
# Define a function to convert centimeters to inches:
def cm_to_inch(value):
  return value/2.54
```

```python
a = cm_to_inch(10)
b = cm_to_inch(15)
fig = plt.figure(figsize=(a,b))
```

```python
names = ["A","B","C"]
values = [19,50,29]
values_2 = [48,19,41]

a = cm_to_inch(10)
b = cm_to_inch(15)
fig = plt.figure(figsize=(a,b))

# Adds subplot on position 1
ax = fig.add_subplot(1,2,1)
# Adds subplot on position 2
ax2 = fig.add_subplot(1,2,2)

ax.bar(names,values)
ax2.bar(names,values_2)
plt.show()
```

### Plots titles

```python
names = ["A","B","C"]
values = [19,50,29]
values_2 = [27,15,34]

fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
# Set the title of the subplot on position 1
ax.set_title("Plot Title")

ax.bar(names, values)
ax2.bar(names, values_2)
# Set the title of the entire figure
plt.suptitle("Test Plots")
plt.show()
```

### Ticks

```python
import matplotlib.pyplot as plt
import numpy as np
import math

range = np.arange(0, math.pi*2, 0.25)
sin = np.sin(range)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.75, 0.75])
ax.plot(range, sin)
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_title("sine")
ax.set_xticks([0,2,4,6,8])
ax.set_xticklabels(["zero","two","four","six","eight"])
ax.set_yticks([-1,0,1])

plt.show()
```

```python
# Grid of 9 subplots
fig, axes = plt.subplots(3,3)

plt.sca(axes[1,1])

plt.xticks([0,1,2],["A","B","C"], color="blue")
plt.yticks([0,1,2],[1,5,10], color="blue")
plt.show()
```

```python
# Direct plotting
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(20,5),columns=["Jan","Feb","March","April","May"])
df.plot.bar(figsize=(20,10)).legend(bbox_to_anchor=(1.1,1))
```

```python
# Stacked bar plot
import pandas as pd
df = pd.DataFrame(np.random.rand(20,5),columns=["Jan","Feb","March","April","May"])
df.plot.bar(stacked=True,figsize=(20,10)).legend(bbox_to_anchor=(1.1,1))
```

```python
# Bar plots
import pandas as pd
df = pd.DataFrame(np.random.rand(20,5),columns=["Jan","Feb","March","April","May"])
df.plot.barh(stacked=True,figsize=(20,10)).legend(bbox_to_anchor=(1.1,1))
```

```python
# Bar's bins Attribute
import pandas as pd
df = pd.DataFrame(np.random.rand(20,5),columns=["Jan","Feb","March","April","May"])
df.plot.hist(bins=20,figsize=(20,10)).legend(bbox_to_anchor=(1.1,1))
```

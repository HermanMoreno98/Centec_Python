## Repetitivas

### For loops

image.png

````python
for i in stock_tickets:
  if i == "GOOGL":
    break
  print(i)
````

````python
stock_tickets = ["AAPL","MSFT","GOOGL",None,"AMZN","CSCO","ORCL"]
stock_tickets
````

````python
### Continue statement ###
# La declaracion de continuar rechaza todas las declaraciones restantes en la iteracion actual del 
# bucle y mueve el control de nuevo a la parte superior del bucle.
for ticket in stock_tickets:
  if(ticket is None):
    continue
    print("-------")
  print(ticket)
````

````python
### Break statement ###
# Proporciona la oportunidad de cerrar un bucle cuando se activa una condición externa
for ticket in stock_tickets:
  if(ticket is None):
    break
    print("-------")
  print(ticket)
````

````python
### Pass statement ###
for ticket in stock_tickets:
  if(ticket is None):
    pass
    print("-------")
  print(ticket)
````

````python
#Conversion de grados celcius a fahrenheit
#Fórmula de conversión de grados Fahrenheit a Celsius:
#(°F – 32) * 5/9 = °C
#Fórmula de conversión de grados Celsius a Fahrenheit:
#(°C * 9/5) + 32 = °F 
fahrenheit = int(input('Ingrese una temperatura en grados Fahrenheit: '))

celsius = (fahrenheit - 32) * 5.0/9.0

# Hacemos uso de la funcion str.format() para darle formato a nuestra respuesta
print('{} grados Fahrenheit son {} grados Celsius '.format(fahrenheit, celsius))
````

````python
def Menu():
  """Funcion que Muestra el Menu"""
  print("""************
  Calculadora
  ************
  Menu
  1) Suma
  2) Resta
  3) Multiplicacion
  4) Division
  5) Salir""")
def Calculadora():
    """Funcion Para Calcular Operaciones Aritmeticas"""
    Menu()
    opc = int(input("Selecione Opcion\n"))
    while (opc >0 and opc <5):
        x = int(input("Ingrese Numero\n"))
        y = int(input("Ingrese Otro Numero\n"))
        if (opc==1):
            print("La Suma es:", x+y)
            opc = int(input("Selecione Opcion\n"))
        elif(opc==2):
            print("La Resta es:",x-y)
            opc = int(input("Selecione Opcion\n"))
        elif(opc==3):
            print("La Multiplicacion es:",x*y)
            opc = int(input("Selecione Opcion\n"))
        elif(opc==4):
            try:
              print("La Division es:", x/y)
              opc = int(input("Selecione Opcion\n"))
            except ZeroDivisionError:
              print("No se Permite la Division Entre 0")
              opc = int(input("Selecione Opcion\n"))
Calculadora()
````

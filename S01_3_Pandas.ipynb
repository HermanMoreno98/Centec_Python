# Missing Values

Variour factors may lead to missing data values

image.png

````python
import pandas as pd
````

````python
first_series = pd.Series([1,2,3,4,5],index=["a","b","c","d","e"])
second_series = pd.Series([10,20,30,40,50],index=["c","e","f","g","h"])
````

````python
sum_of_series = first_series + second_series
sum_of_series
````

````python
# drop NaN(Not a Number) values from dataset
dropna_s = sum_of_series.dropna()
dropna_s
````

````python
# fill NaN values with Zeros(0)
fillna_s = sum_of_series.fillna(0)
fillna_s
````

````python
# fill values with zeros before performing addition operation for missing indices
fill_NaN_with_zeros_befores_sum = first_series.add(second_series, fill_value=0)
fill_NaN_with_zeros_befores_sum
````

# Data Operation

````python
df_movie_rating = pd.DataFrame({
    "movie_1":[5,4,3,3,2,1],
    "movie_2":[4,5,2,3,4,2]},
    index=["Tom","Jeff","Peter","Ram","Ted","Paul"]
)
df_movie_rating
````

````python
def movie_grade(rating):
  if rating==5:
    return "A"
  elif rating==4:
    return "B"
  elif rating==3:
    return "C"
  else:
    return "F"
````

````python
#Test the function
print(movie_grade(5))
````

````python
#Apply the function to the dataframe
df_movie_rating.applymap(movie_grade)
````

````python
df_movie_rating.movie_1.apply(lambda x: movie_grade(x))
````

````python
#List comprehension
[movie_grade(x) for x in df_movie_rating['movie_1']]
````

````python
for x,y in zip(df_movie_rating['movie_1'],df_movie_rating['movie_2']):
  print(movie_grade(x))
  print(movie_grade(y))
````

````python
[(movie_grade(x),movie_grade(y)) for x,y in zip(df_movie_rating['movie_1'],df_movie_rating['movie_2'])]
````

# Data Standardization

````python
#Create a function to return the standardize value
def standardize_test(test):
  return (test-test.mean()) / test.std()
````

# Pandas SQL Operations

````python
import pandas as pd
import sqlite3
````

````python
#Create SQL table
create_table = """
CREATE TABLE student_score
(Id INTEGER, Name VARCHAR(20), Math REAL, Science REAL);
"""
````

````python
#execute the SQL statement
executeSQL = sqlite3.connect(":memory:")
executeSQL.execute(create_table)
#Save the changes in the database
executeSQL.commit()
````

````python
#prepare a SQL query
SQL_query = executeSQL.execute("SELECT * FROM student_score")
````

````python
#fetch result from the SQLlite database
resultset = SQL_query.fetchall()
````

````python
#view result (empty data)
resultset
````

````python
#prepare records to be inserted into SQL table through SQL statement
insertSQL = [(10,"Jack",85,92),
             (29,"Tom",73,89),
             (65,"Ram",65.5,77),
             (5,"Steve",55,91)]
````

````python
#insert records into SQL table through SQL statement
insert_statement = "INSERT INTO student_score VALUES (?,?,?,?)"
executeSQL.executemany(insert_statement,insertSQL)
executeSQL.commit()
````

````python
#prepare SQL query
SQL_query = executeSQL.execute("SELECT * FROM student_score")
````

````python
#fetch the resulset for the query
resulset = SQL_query.fetchall()
````

````python
#View the resultset
resulset
````

````python
#put the records together in dataframe
df_student_recors = pd.DataFrame(resulset,columns=list(zip(*SQL_query.description))[0])
````

````python
#view the records in pandas dataframe
df_student_recors
````

````python
# Consultar los registros de la tabla "stocks" donde el precio es mayor a 100
cursor = executeSQL.cursor()
SQL_query = cursor.execute("SELECT * FROM student_score WHERE Math > 67")
````

````python
# Leer los resultados de la consulta
results = cursor.fetchall()
results
````
